```python
# 1. Zaimportowane biblioteki

import numpy as np
# import scipy # nieużywane
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
# DataFrame z modułu pandas
# import sklearn # nieużywane
import seaborn as sns

# 2. Wczytane dane

df = pd.read_csv("winequality-red.csv", sep=';')
df.head()
```

## 3. Analiza struktury zbioru

```python
# 3.a) Rozmiary zbioru (l. wierszy, l. kolumn)

df.shape
```

```python
# 3.b) Typy zmienych

df.dtypes
```

```python
# 3.c) Liczba próbek (wierszy) w każdej klasie wyniku "quality"

output = "quality"

print(df[output].value_counts(), '\n')
```

## 4. Brakujące dane

```python
# 4. Wykrywanie brakujących wartości

nulls = pd.DataFrame(df.isnull().sum(), columns=['Nulls'])
print(nulls)
```

## 5. Podsumowanie zbioru

```python
# 5. Podsumowanie zbioru (l. elementów, średnia, odchylenie stand., ...)
#    dla trzech zmiennych

variables = ["alcohol", "pH", "residual sugar"]
df[variables].describe()
```

## 6. Wykresy

```python
# 6.a) Wykresy liniowe (/punktowe)

for var in variables:
#     df[var].plot()
#     plt.ylabel(var)
    sns.scatterplot(data=df[var])
    plt.show()
```

```python
# 6.b) Histogramy

for var in variables:
    sns.displot(df[var])
    plt.show()
```

```python
# 6.c) Wykresy pudełkowe (wzg. klasy "quality")

for var in variables:
    sns.catplot(data=df,
                x=output,
                y=var,
                kind="box")
    plt.show()
```
