```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
print(X)
```

```python
y = np.dot(X, np.array([1, 2])) + 3
print(y)
```

```python
reg = LinearRegression().fit(X, y)
```

```python
X_t = np.array([[3, 3], [3, 4], [4, 5]])
y_t = np.array([12, 14, 18])
reg.score(X_t, y_t)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(np.array([[4, 4]]))
```

## scikit-learn

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X, y = datasets.load_diabetes(return_X_y=True)

import pandas as pd
data = pd.read_csv("winequality-red.csv", sep=";")
data.head()
```

```python
variable = "fixed acidity"
output = "pH"

X = data[[variable]]
y = data[output]

# X = X[:, np.newaxis, 2]
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.show()
```

```python
test_obj_amount = 200
X_train = X[:-test_obj_amount]
X_test = X[-test_obj_amount:]

y_train = y[:-test_obj_amount]
y_test = y[-test_obj_amount:]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
print(f"Coefficient: {regr.coef_}")
print(f"Mean squared error: {mean_squared_error(y_test, y_pred)}")
print(f"Coefficient of determination: {r2_score(y_test, y_pred)}")
```

```python
plt.scatter(X_train, y_train, color="black", alpha=0.1)
plt.scatter(X_test, y_test, color="red")
plt.scatter(X_test, y_pred, color="blue")

# plt.xticks(())
# plt.yticks(())

plt.xlabel(variable)
plt.ylabel(output)
# plt.axis([0, 10, None, None])

plt.show()
```
